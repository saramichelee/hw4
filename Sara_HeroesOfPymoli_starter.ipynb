```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data["Item ID"].nunique()
unique_items
```

```python
average_price = purchase_data["Price"].mean()
average_price
```

```python
purchase_count = purchase_data["Purchase ID"].count()
purchase_count
```

```python
total_revenue = purchase_data["Price"].sum()
total_revenue
```

```python
stats = {"Number of Unique Items" : [unique_items],
        "Average Price" : [average_price],
        "Number of Purchases" : [purchase_count],
        "Total Revenue" : [total_revenue]}

stats_df = pd.DataFrame(stats)
stats_df
```

```python
stats_df["Average Price"] = stats_df["Average Price"].map("${:.2f}".format)
stats_df["Total Revenue"] = stats_df["Total Revenue"].map("${:.2f}".format)
stats_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

age_range = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

unique_df["Age Range"] = pd.cut(unique_df["Age"], bins, labels=age_range)
unique_df.head()
```

```python
age_group = unique_df.groupby("Age Range")
age_count = pd.DataFrame(age_group["SN"].count())
age_count["Age Percentage"] = round(age_count["SN"]/total_players*100,2)
age_count = age_count.rename(columns = {"SN" : "Age Count"})
age_count
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
data_df["Age Range"] = pd.cut(data_df["Age"], bins, labels=age_range)
data_df.head()
```

```python
age_unique = pd.DataFrame(data_df.groupby("Age Range")["SN"].count())
age_unique
```

```python
age_unique["Average Purchase Price"]= data_df.groupby("Age Range")["Price"].mean().astype(float)
age_unique["Total Purchase Value"] = data_df.groupby("Age Range")["Price"].sum().astype(float)
age_unique
```

```python
range_unique = pd.DataFrame(data_df.groupby("Age Range")["SN"].nunique())
range_unique
age_unique["Avg Total Purchase per Person"] = age_unique["Total Purchase Value"]/range_unique["SN"].astype(float)
age_unique
```

```python
age_unique["Average Purchase Price"] = age_unique["Average Purchase Price"].map("${:.2f}".format)
age_unique["Total Purchase Value"] = age_unique["Total Purchase Value"].map("${:.2f}".format)
age_unique["Avg Total Purchase per Person"] = age_unique["Avg Total Purchase per Person"].map("${:.2f}".format)
age_unique = age_unique.rename(columns={"SN" : "Purchase Count"})
age_unique
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
player_pivot = data_df.groupby("SN")
purchase_count = pd.DataFrame(player_pivot["Item ID"].count())
purchase_count["Average Purchase Price"]= pd.DataFrame(player_pivot["Price"].mean().astype(float))
purchase_count["Total Purchase Value"]= pd.DataFrame(player_pivot["Price"].sum().astype(float))
purchase_count.head()
```

```python
highest = purchase_count.sort_values(by='Item ID', ascending=False)
highest.head(10)
```

```python
highest_format = highest.rename(columns = {"Item ID" : "Purchase Count"})
highest_format = highest_format.sort_values(by='Total Purchase Value', ascending=False)
highest_format["Average Purchase Price"] = highest_format["Average Purchase Price"].map("${:.2f}".format)
highest_format["Total Purchase Value"] = highest_format["Total Purchase Value"].map("${:.2f}".format)
highest_format.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items = data_df[["Item ID","Item Name","Price"]]
items.head(10)
```

```python
item_group = pd.DataFrame(items.groupby(['Item ID', 'Item Name'])['Price'].agg(['count','mean','sum']))
item_group.head()
```

```python
high_purch = item_group.sort_values(by='count', ascending=False)
high_purch.head()
```

```python
item_format = high_purch.rename(columns={"count" : "Purchase Count", "mean" : "Item Price",
                               "sum" : "Total Purchase Value"})
item_format["Item Price"] = item_format["Item Price"].map("${:.2f}".format)
item_format["Total Purchase Value"] = item_format["Total Purchase Value"].map("${:.2f}".format)
item_format.head()
```
